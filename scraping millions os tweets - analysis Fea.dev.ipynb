```python
# Importing libraries and packages
import snscrape.modules.twitter as sntwitter
import pandas as pd
import time
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
start_time = time.time()

# Creating list to append tweet data 
tweets_list1 = []

#Joaozin_A
# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('from:Joaozin_A').get_items()): #declare a username 
    if i>1000: #number of tweets you want to scrape
        break
    #declare the attributes to be returned
    tweets_list1.append([tweet.date, 
                         tweet.id, 
                         tweet.content, 
                         tweet.source, 
                         tweet.user.username, 
                         tweet.media,
                         tweet.lang,
                         tweet.user.verified,]) 
    
# Creating a dataframe from the tweets list above 
teste = pd.DataFrame(tweets_list1, columns=['Datetime',
                                                 'Tweet Id',
                                                 'Text',
                                                 'Source',
                                                 'Username',
                                                 'Media',
                                                 'language',
                                                 'verified'])

end_time = time.time()
print(f"Result calculated in {end_time - start_time} sec")
```

```python
start_time = time.time()

# Creating list to append tweet data to
tweets_list2 = []

# Using TwitterSearchScraper to scrape data and append tweets to list > O termo pode ser composto. ex: COVID Vaccine
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('@fea_dev since:2021-01-01 until:2022-05-31').get_items()):
    if i > 5000:
        break
    tweets_list2.append([tweet.date, 
                         tweet.id, 
                         tweet.content, 
                         tweet.user.username,
                         tweet.mentionedUsers,
                         tweet.retweetedTweet,
                         tweet.likeCount,
                         tweet.retweetCount,
                         tweet.user.location,
                         tweet.user.followersCount,
                         tweet.user.verified])
    
# Creating a dataframe from the tweets list above
dev = pd.DataFrame(tweets_list2, columns=['Datetime', 
                                          'Tweet Id', 
                                          'Text', 
                                          'Username',
                                          'MentionedUsers',
                                          'retweetedTweet',
                                          'Likes',
                                          'Retweets',
                                          'Location',
                                          'Followers',
                                          'Verified'])

end_time = time.time()
print(f"Result calculated in {end_time - start_time} sec")
```
